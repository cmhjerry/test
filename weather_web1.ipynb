```python
column_name = []
column_name.append("date")
for i in range (8):
    column_name.append(line_list[i+1])
column_name.append("Holiday_Weekend")
print(column_name)
```

```python

Date_list = [] 
Temp_list = []
Dew_list = [] 
Humidity_list = [] 
SeaLeverlPress_list = []
Visibility_list = []
Wind_list = []
Precip_list = []
Events_list = []


StartDate = "01/01/15"
Date_start = datetime.datetime.strptime(StartDate, "%m/%d/%y")
#########################get the first date beginning index###################
index = 0
#find the first index:
for line in line_list:  
    if line == '1':
        break
    index = index + 1
print(index)
print(line_list[index])

############################################
for i in range(365):
    #get the data
    
    Date_string = '{:%m/%d/%Y}'.format(Date_start)
    print(Date_string)
    Date_day = str(int(Date_string.split('/')[1]))
    print(Date_day)
    Date_list.append(Date_string)
    Date_start = Date_start + datetime.timedelta(days=1)
    
    print(line_list[index])
    #get and make sure start index
    while 1:
        if(line_list[index] == Date_day):
            break
        index = index +1
    
    index = index + 2
    Temp_list.append(line_list[index])
    index = index + 3
    Dew_list.append(line_list[index])
    index = index + 3
    Humidity_list.append(line_list[index])
    index = index + 3
    SeaLeverlPress_list.append(line_list[index])
    index = index + 3
    Visibility_list.append(line_list[index])
    index = index + 2
    Wind_list.append(line_list[index])
    index = index + 3
    Precip_list.append(line_list[index])
    index = index + 1
    
    if(line_list[index] == '\t\xa0'):
        Events_list.append("Good")
        index = index + 1
    else:
        temp_weather = line_list[index]
        while 1:
            index = index+1
            if(line_list[index] == '\t,'):
                index = index+1
                temp_weather = temp_weather + "," + line_list[index]
            break
        Events_list.append(temp_weather)
```

```python
#holiday label weekend and friday as 1, Monday to Thursday as 0, Holiday as 2, updated the weekend close to Holiday as 3
Federal_Holidays = ["01/01/2015","01/19/2015","02/16/2015","05/25/2015","07/03/2015","09/07/2015","10/12/2015","11/11/2015","11/26/2015","12/25/2015"]
# 01/01/2017 is Thursday, so set up the start point to 4
start = 4
Holiday_list = []
for i in Date_list:
    if i in Federal_Holidays:
        Holiday_list.append(2)
        start = start+1
        continue
    if start%7 == 5 or start%7 == 6 or start%7 == 0:
        Holiday_list.append(1)
    else:
        Holiday_list.append(0)
    start = start+1
    
index_day = 0 
for i in Holiday_list:
    if i == 2:
        interval_day = 1
        while 1:
            if Holiday_list[index_day+interval_day] == 1:
                Holiday_list[index_day+interval_day]=2
                interval_day = interval_day +1
            else:
                break
    if i == 2 and index_day != 0:
        interval_day = 1
        while 1:
            if Holiday_list[index_day-interval_day] == 1:
                Holiday_list[index_day-interval_day]=2
                interval_day = interval_day +1
            else:
                break
    index_day = index_day + 1  

print(len(Holiday_list))
print(Holiday_list)
```

## update events to be numerable

```python
print(type(Events_list[0]))

# good 1
# rain 10
# snow 100
# fog 1000
# Thunderstorm 10000

for i in range(365):
    temp = 0
    if "Good" in Events_list[i]:
        temp=1
    if "Thunderstorm" in Events_list[i]:
        temp=temp+10000
    if "Fog" in Events_list[i]:
        temp=temp+1000
    if "Snow" in Events_list[i]:
        temp=temp+100
    if "Rain" in Events_list[i]:
        temp=temp+10
    Events_list[i] = temp
```

```python
import csv
result = []
for i in range(365):
    line = []
    line.append(Date_list[i])
    line.append(Temp_list[i])
    line.append(Dew_list[i])
    line.append(Humidity_list[i])   
    line.append(SeaLeverlPress_list[i])
    line.append(Visibility_list[i])
    line.append(Wind_list[i])
    line.append(Precip_list[i])
    line.append(Events_list[i])
    line.append(Holiday_list[i])
    result.append(line)

with open("2015BOSweather.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(column_name)
    writer.writerows(result)
    
```
